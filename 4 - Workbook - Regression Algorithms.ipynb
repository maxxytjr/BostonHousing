## I. The "Final Boss" of Machine Learning

```python
import numpy as np

import pandas as pd
pd.set_option('display.max_columns', 100)

from matplotlib import pyplot as plt
%matplotlib inline 

import seaborn as sns
sns.set_style('darkgrid')
```

```python
x = np.linspace(0, 2*np.pi, 100)

# noise
np.random.seed(321)
noise = np.random.normal(0, 0.5, 100)

# target
y = np.sin(x) + noise
```

```python
df = pd.DataFrame({'x': x,
                  'y': y})
df.head()
```

```python
plt.scatter(df.x, df.y)
plt.plot(df.x, np.sin(df.x), 'k--')
plt.show()
```

## II. Flaws of Linear Regression

**Simple linear regression suffers from two major flaws:**

1. It's prone to overfitting with many input features.
2. It cannot easily express non-linear relationships.

```python
from sklearn.linear_model import LinearRegression
```

```python
# Initialize model instance
lm = LinearRegression()

# Separate input feature(s) and target variable
X = df.drop('y', axis=1)
y = df.y

# Fit model
lm.fit(X, y)
```

```python
# Scatterplot of x and y
plt.scatter(df.x, df.y)

# Overlay the sine wave
plt.plot(df.x, lm.predict(X), 'k--')

plt.show()
```

### Two ways to Increase Model Complexity:

1. Inherent structure of the model. For egs, decision trees and polynomial regression models are more complex and can represent non-linear relationships

2. Increase the number of input features (overfitting occurs very easily with linear regression models when we add features)

## III. Regularized Regression

**Regularization (in the context of linear regression) is a technique used to prevent overfitting by artificially penalizing model coefficients.**

1. It does so by adding a **penalty** to the cost function.
2. This dampens coefficients, or removes the feature entirely, depending on the type of penalty used
3. The 'strength' of the penalty factor is **tunable**

**There are 2 types of penalties**

1. **L1** penalizes the absolute size of the model coefficients
2. **L2** penalizes the squared size of the model coefficients

**Three types of regularization algorithms to be introduced here:**:

1. Lasso
2. Ridge
3. Elastic-Net

**Lasso relies on the L1 regularization penalty**
- This leads to coefficients that can be exactly 0 (**automatic feature selection**)
- The lambda (strength of the penalty) should also be **tuned**. A stronger penalty leads to more coefficients pushed to 0

```python
from sklearn.linear_model import Lasso
```

**Ridge Regression relies on L2 penalty**

- Completely relies on the L2 (squared) penalty
- leads to smaller coefficients but does not turn them to 0 (**feature shrinkage**)
- The strength of the penalty should also be tuned

```python
from sklearn.linear_model import Ridge
```

**Elastic-Net is a compromise between Lasso and Ridge**

- Combines both Lasso and Ridge
- **ratio of L1 and L2** should be tuned
- **strength of penalties** should be tuned
- Ridge and Lasso can be thought of as subsets of elastic-nets if you set the ratio to be biased completely towards either

```python
from sklearn.linear_model import ElasticNet
```

## IV. Ensemble Methods

Earlier in this course we discovered that unconstrained decision trees are prone to overfitting, because they can keep growing until each leaf memorizes an individual observation.

A way to take advantage of this flexibility without overfitting will be to utilize **Tree Ensembles**.

**Random forests train a large number of strong decision trees and combine their predictions through BAGGING**.

**Bagging** attempts to reduce the chance of overfitting complex models. It trains a large number of strong decision trees in parallel. 

A strong model is allowed to have high complexity. The ensemble then combines all these to "smooth out" the predictions.

It does this by:
1. having each decision tree choose a random number of features to train on
2. having each decision tree choose a subset of the observations to train on

**In practice, ensemble methods work very well. They are the perfect swiss army knife algorithm you can count on. They don't have many parameters to tune too**

```python
from sklearn.ensemble import RandomForestRegressor
```

**Boosted trees train a sequence of 'weak', *constrained* decision trees and combines their predictions through *boosting***

Each decision tree is allowed a **maximum depth**, which should be tuned

**In practice, boosted trees tend to have very high performance and can beat many other types of models. They are harder to tune compared to random forests**

```python
from sklearn.ensemble import GradientBoostingRegressor
```

#### C.) Loop through the following 5 candidate values for penalty strength. Fit and plot Lasso Regressions on the garbage dataset using each one.
* Candidate values: <code>[0.01, 0.05, 0.1, 0.5, 1]</code>.
* Label each plot's title as **Alpha: penalty** where "penalty" is the dynamically inserted penalty value used to train that model.
* Based on a quick eyeball test, which one appears to have the best performance on the garbage dataset?

What you just did could be considered a very rudimentary form of **model tuning.** You tried out different candidate values for an important hyperparameter (penalty strength), evaluated their performances, and made an initial guess on the optimal value for that hyperparameter.

As we'll see in the next module, the actual model tuning process will not be much different in terms of setup, but it WILL be much more systematic, automated, and empirical. Specifically, it will incorporate **hyperparameter grids** and **cross-validation** to help you reliably and automatically arrive at the best hyperparameter values.

## <span style="color:RoyalBlue">Exercise 4.3 - Ridge Regression</span>

Whereas Lasso regression relies on the $L_1$ penalty, Ridge regression completely relies on the $L_2$ penalty (**squared**) instead. Practically, this leads to smaller coefficients, but it doesn't force them to 0.

#### A.) Fit and plot a Ridge Regression on the garbage dataset with <code>alpha=1.0</code> and <code>random_state=1234</code>.
* Is this overfit or underfit?

```python
ridge = Ridge(random_state=1234, alpha=1.0)
ridge.fit(X_garbage, y)

plt.scatter(df.x, df.y)
plt.plot(df.x, ridge.predict(X_garbage), 'k--')
plt.show()
```

**Overfit! Decrease overfitting by increasing the penalty strength (L2; feature dampening)**

#### B.) Based on your plot from part (A) and what you know about penalty strengths, what do you think will happen if you reduce penalty strength to <code>alpha=0.5</code>?
* Go ahead and try it out. Fit and plot the Lasso Regression with <code>alpha=0.5</code>.

```python
ridge = Ridge(random_state=1234, alpha=0.5)
ridge.fit(X_garbage, y)

plt.scatter(df.x, df.y)
plt.plot(df.x, ridge.predict(X_garbage), 'k--')
plt.show()
```

#### C.) Based on what you've learned so far, propose 5 candidate values for the penalty strength of the Ridge Regression.
* Why did you choose those 5 values?
* Fit and plot them to evaluate their performance.
* See the answer key for the 5 values we chose (there's no single "right" answer).

```python
alphas = [1, 10, 100, 1000, 10000]

for alpha in alphas:
    ridge = Ridge(random_state=1234, alpha=alpha)
    ridge.fit(X_garbage, y)
    
    plt.scatter(df.x, df.y)
    plt.plot(df.x, ridge.predict(X_garbage), 'k--')
    plt.title("Penalty: {}".format(alpha))
    plt.show()
```

#### D.) What do you think happens if you increase the penalty strength all the way up?
